# Boxplots:
#	Create boxplots for numerical variables to identify outliers and the interquartile range.

```python

plt.boxplot(df['Total Sales Value'])
plt.show()
```

```python
# Plot the Boxplot
sns.boxplot(x=df['Total Sales Value'])
plt.title("Boxplot to Visualize Outliers")
plt.show()

# Identify Outliers Programmatically
Q1 = df['Total Sales Value'].quantile(0.25)  
Q3 = df['Total Sales Value'].quantile(0.75) 
IQR = Q3 - Q1  # Interquartile Range

print(' First Quartile',Q1)
print(' Third Quartile',Q3)
print("IQR",IQR)


# Outlier Range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("lower_bound")
print("upper_bound")

# Filter Outliers
outliers = df[(df['Total Sales Value'] < lower_bound) | (df['Total Sales Value'] > upper_bound)]
print("Outliers:", outliers)
```

```python
plt.boxplot(df['Discount Rate (%)'])
plt.show()


# Plot the Boxplot
sns.boxplot(x=df['Discount Rate (%)'])
plt.title("Boxplot to Visualize Outliers")
plt.show()

# Identify Outliers Programmatically
Q1 = df['Discount Rate (%)'].quantile(0.25)  
Q3 = df['Discount Rate (%)'].quantile(0.75) 
IQR = Q3 - Q1  # Interquartile Range

print(' First Quartile',Q1)
print(' Third Quartile',Q3)
print("IQR",IQR)


# Outlier Range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("lower_bound")
print("upper_bound")

# Filter Outliers
outliers = df[(df['Discount Rate (%)'] < lower_bound) | (df['Discount Rate (%)'] > upper_bound)]
print("Outliers:", outliers)
```
